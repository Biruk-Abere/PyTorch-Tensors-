# Tensors In Physics : A Brief Reminder 
Let's recall how tensors helped the 19th century physics. Physics starts with measuring and describing the values of different physical quantities. It goes on to equations which enable us to predict the values of this quantities. 

A measuring instrument usually returns a single numerical value. For some physical quantities like mass m , the single measured value is sufficent to describe the quantity. For other quantities , we need several values. For example we need three components Ex , Ey and Ez to describe the electric field at a given point. To describe the tension inside a solid body , we need even more values , we need 6 values. to describe the tension inside a solid body. 

The problem was that in the 19 century , physicists used a separate equation for each component of the field. As a result , equations were cumbersome and difficult to solve. 

The main idea of the tensor approach is to describe all of the components of a physical field as a single mathematical object. 
  

*   A vector ai , 
*    or more generally , a tensor aij , aijk , .....

As a result , we got simplified equations and faster computations. 

It is worth mentioning that originally, most vectors (rank-1 tensors) were used. However the 20 century physics has shown the higher - order matrices are also useful. For example 


*   Matrices (rank-2 tensors ) are actively used in quantum physics 
*   Higher order tensors such as the rank-4 curvature tensor are actively used in the General Relativity Theory 

So the word Tensor means streching someting , in this case streching of numbers , because Tensors are a datastructures of n-dimensional matrices. It's a mathematical object that is analogues to but more general than a vector , represented by an array of components that are functions of the coordinates of a space.


In math , any kind of n-dimensional matrix is called Tensors


*  1*1 scalars are tensors 
*  1*n vectors are scalars
*  n*n matrices are tensors
*  n*n*n multi dimensional arrays are also tensors

- In Python deep learning model , all of our data , inputs outputs learning weights these are going to be expressed as tensors , multi dimensional arrays that can contain floating point , integer or boolean data.

# Creating Tensors 

The simplest way to create a tensor is using torch.empty

```python
import torch
empty_tensor = torch.empty(3,2)
print(type(empty_tensor))
print(empty_tensor)
```

Let's unpack what we just did :- 
- We created a tensor using one of the numbers factory methods attached to the torch module
- The tensor itself is a 2-dimensional having 3 rows and 4 columns 
- The type of object returned is torch.tensor , which is an alias from torch.FloatTensor by default. 
- We will probably see random-looking values when printing our Tensor. The torch.empty() call allocates memory for the tensor but does not initialize it with any values so what we're seeing is whatever was in memory at the time of allocation. 

More often that not , we'll want to initialize with some value. common cases are all zeros , all ones or random values and the torch module provides factory methods for all of these

```python
import torch

ones_tensor = torch.ones(2 , 3)
print(ones_tensor)

zeros_tensor = torch.zeros(2 , 3)
print(zeros_tensor)

torch.manual_seed(1729)
random_tensor = torch.rand(2 , 3)
print(random_tensor) # full of random values between 0 and 1
```

# Tensor Shapes 

Often when we're performing operations one two or more tensors they will need to be the same shape that is having the same number of dimensions and the same number of cells in each dimension. For that we have torch._like methods. 

```python
import torch
x = torch.empty(2 , 2 , 3 )
print(type(x))
print(x.shape)
print(x)

x_empty = torch.empty_like(x)
print(x_empty.shape)
print(x_empty)

x_zeros_like = torch.zeros_like(x)
print(x_zeros_like.shape)
print(x_zeros_like)

x_ones_like = torch.ones_like(x)
print(x_ones_like.shape)
print(x_ones_like)

x_rand_like= torch.rand_like(x)
print(x_rand_like.shape)
print(x_rand_like)
```

# Tensor Datatypes 

setting the datatype of a tensor is possible in couple of ways:

```python
import torch
tensor_vector = torch.tensor([1 , 2 , 3 , 4 ])
print(tensor_vector.dtype)

# there are a couple of ways to chnage a data type from one to another 
# We can change at the time of declaration
tensor_matrix =  torch.tensor([[1 , 2 , 3 , 4 ] , [5 , 6 , 7 , 8]] , dtype = torch.int16)
print(tensor_matrix.dtype)

# we can change using .to() method
tensor32 = tensor_matrix.to(torch.float64)
print(tensor32.dtype)
```

The simplest way to set the underlying datatype of a tensor is with an optional argument at creation time. In the first line of the cell above , we set a datatype of dtype = torch.int16 for the tensor , tensor_matrix. When we print tensor matrix it will have a datatype of int16. PyTorch will take a serious of initial , unlabeled integer arguments as a tensor shape but when adding the optional arguments . It can make the intent more readable. 

- The other way to set a datatype is with the .to() method. We create tensor32 by converting tensor_matrix to a 64bit float with the .to method.

# Copying A Tensor

As with any object in python assigning a tensor to a variable makes the variable a label of the tensor and does not copy it , for example 

```python
import torch
a = torch.ones(2,2)
b = a # assigning the value of a to the b
print(torch.eq(b , a))
a[0][1] = 20
print(b)
```

But what if we want a separate copy of the data to work on ? the clone method is there for you. 

```python
import torch
a = torch.ones(2,2)
b =  a.clone() # clone the value of a to b but in another memory
assert b is not a # check that the memory slot b is not in the memory slot of a
a[0][1] = 20 # a changes
print(b) # but b is still all ones 
```

The assert keyword is used when debugging a code , the assert keyword let's our test if a condition in our code returns True , if not the program will raise Assertion Error. In our case we want to assert b and a are different objects in memory.

# Moving To GPU

One of the major advantages of PyTorch is its robust acceleration on CUDA compatible NVDIA GPU's (CUDA stands for compute unified device architecture) which is NIVIDIA'S platform for parallel computing ? 

So far everything we've done has been on CPU. How do we move to the faster hardware ? 

First we should check whether a GPU is available , with the is_available() method. 

Note :- If we do not have a CUDA compatible GPU and CUDA drivers installed , the executable cells in this section will not execute any GPU related code.  

```python
import torch
if torch.cuda.is_available():
  print("GPU is available")

else :
  print("GPU is not available")
  
```

Once we've determined that one or more GPU's available. we need to put our data someplace where the GPU can see it. Our CPU does computation in our computer's RAM , our GPU has dedicated memory detached to it. Whenever we want to perform a computation on a device , we must move all the data needed for that computation to memory accessible by that device.  

There are multiple ways to get our data on to our target device. We may do it at a creation time. 

```python
import torch
if torch.cuda.is_available():
  random_cuda = torch.rand((2 ,2) , device = 'cuda')
  print(random_cuda)
else :
  print("sorry cpu only")
```

By default , new tensors are created on the CPU , so we have to specify when we want to create our tensor on the GPU  with the optimal device argument. we can see when we print the new tensor , PyTorch informs us which device it's on (if it is not on the cpu)


- We can query the number of the GPU's with torch.cuda.device_count() , if we have more than one GPU , we can specify them by index , device = 'cuda:0' , device='cuda:1'etc...

- As a coding practice specifying our devices everywhere with string constraints is pretty fragile , in an ideal world , our code would perform robustly wheather we are on CPU or GPU hardware. we can do this by creating a device handle that can be passed to our tensors instead of a string. 

```python
import torch
if torch.cuda.is_available():
  print(torch.cuda.device_count())
else:
  print("Sorry , CPU only")
```

```python
import torch
if torch.cuda.is_available():
  my_device = torch.device('cuda')
else:
  my_device = torch.device("CPU")

x = torch.rand((2,3) , device = my_device)
print(x)
```

If we have an existing tensor living on one device , we can move it to another with the to() method. The following line of code creates a tensor on the CPU and moves it to which ever device handle we acquired in the previous cell. 

```python
import torch
y = torch.tensor([1 , 2 , 3 , 4])
y = y.to(my_device)
```

It is important to know that inorder to do computation involving two or more tensor , all of the tensors must be on the same device. The following code will throw a run time error regardless of wheather we have a GPU device available. 

```python
import torch
tensor = torch.tensor([1 ,2 ,3 , 4])
y = torch.rand([4 , 5 , 6 , 7] , device = 'my_device')
z = y+tensor
print(z) # exception will be thrown
```

## Manipulating Tensor Shapes 

# Change The Number Of Dimensions

one case where we might need to change the number of dimensions is passing a single instance of input in to our model. PyTorch models generally expects batches of input. 

Forexample:- Imagine having a model that works on 3*226*226 images , a 226 pixel square with a 3 color channels , when we load and transform it , we will get a tensor of shape (3,226,226) but our model through is expecting input shape (N , 3 , 226 , 226) where N is the number of images in the batch. 

```python
import torch
a = torch.rand(3,226,226)
b = torch.unsqueeze(a , 0)
print(a.shape)
print(b.shape)
```

```python
import torch
a = torch.rand(3 , 226 , 226)
b = a.unsqueeze(0)
print(a.shape)
print(b.shape)
```

The unsqueeze() method adds a dimension of extent 1. Unsqueeze(0) adds it as a new zeroth dimension. so if that's unsqueezing ? what do we mean by squeezing ? 

Continuing the example above , let's say the model's output is a 20 element vector for each input. We would then expect the output to have shape (N,20) , where N is the number of instances in the input batch. That means that for our single input batch we'll get an output shape (1,20). what if we want to do some non-batched computation with that output - something that's just expecting a 20-element vector. 

```python
import torch
a = torch.rand(1 , 20)
b = torch.squeeze(a , 0)
print(a.shape)
print(b.shape)
```

we can see from the shapes that our 2-dimensional tensor is now 1 dimensional

Sometimes we'll want to change the shape of a tensor more radically , while still preserving the number of elements and their contents Sometimes we'll want to change the shape of a tensor more radically , while still preserving the number of elements and their contents. One case where this happens is at the interface between a convolutional layer of a model and a linear layer of the model. This is common in image classification. A convolutional kernel will yield an output tensor of shape features X width X height but the following linear layer epects a 1 dimenisonal input.reshape() will do this for you , provided that the dimensions we request yield the same number of elements as the input tensor has:-

```python
import torch
a = torch.rand(3 , 226 , 226)
b = a.reshape(3*226*226)
print(a.shape)
print(b.shape)
```

# Numpy Bridge 

In this section above on broadcasting it was mentioned that PyTorch's broadcast semantics are compatible with numpy's but the kinship between PyTorch and Numpy goes even deeper than that.

If we have existing ML or scientific code with data stored in Numpy ndarrays , we may wish to express what some data as pytorch tensors , wheather to take advantage of PyTorch's GPU acceleration, or it's effecient abstractions for building ML models. it's easy to switch between ndarrays and PyTorch tensors. 

```python
import torch
import numpy as np
# converting from numpy to tensor
numpy = np.array([[1 , 2 , 3 , 4 ] , [5 , 6 , 7 , 8]])
from_numpy_to_tensor = torch.from_numpy(numpy)
print(from_numpy_to_tensor.type)

# Converting from tensor to numpy
tensor_to_numpy = torch.tensor([1 , 2 , 3 ,4 ])
yes_tensor_to_numpy = tensor_to_numpy.numpy()
print(type(yes_tensor_to_numpy))
```
